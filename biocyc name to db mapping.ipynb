```python
import requests
import json
from xml.etree import ElementTree as ET

with open('..\\dissertation DLC content\\go away\\passwords.json', 'r') as f:
    passwords = json.load(f)
    password = passwords.get('metacyc_pw')

email = 'lp2095@bath.ac.uk'

session = requests.Session()

databases = session.get(f'https://websvc.biocyc.org/xmlquery?dbs')
```

```python
#e.g. Saccharomyces cerevisiae -> YEAST
db_mapping = {}
```

```python
dbs_xml = ET.fromstring(databases.text)

dbs = dbs_xml.find('metadata').findall('PGDB')

for db in dbs:
    db_id = db.get('orgid')
    species_name = db.find('species').text.replace('[', '').replace(']', '')
    strain = db.find('strain')
    if strain != None:
        strain = strain.text
    else:
        strain = db.find('subspecies')
        if strain != None:
            strain = strain.text
        else:
            strain = ' '

    if not db_mapping.get(species_name):
        db_mapping.update({species_name: {}})
        
    db_mapping[species_name][strain] = db_id
    
print(db_mapping)
```

```python
with open('list_files\\biocyc_db_mappings.json', 'w', encoding='utf-8') as f:
    json.dump(db_mapping, f)
```
