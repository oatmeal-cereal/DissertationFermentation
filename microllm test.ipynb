```python
from transformers import AutoTokenizer as autok
from transformers import BertForTokenClassification as bert
#from transformers import BertForSequenceClassification as bert
from transformers import AutoModel as autom
from transformers import AutoModelForCausalLM as causallm
from transformers import pipeline
%pip install peft
from peft import get_peft_model, LoraConfig, TaskType

from huggingface_hub import notebook_login
```

```python
chatglm = 'chatglm3-6b'

model = causallm.from_pretrained(chatglm, trust_remote_code=True)
```

```python
peft_config = LoraConfig.from_pretrained(model, task_type=TaskType.CAUSAL_LM)
peft_model = get_peft_model(model, peft_config)

nlp = pipeline('causal-lm', model=peft_model, aggregation_strategy='first')
```
