```python
import spacy
from spacy.pipeline import SpanRuler
from spacy.tokens import Doc
import re
import json
```

```python
model = spacy.load('en_core_web_sm')
ruler = model.add_pipe('span_ruler', before='ner')

patterns = [
    [{'TEXT': {'REGEX': '[A-Za-z]+'}}],
    [{'TEXT': 'hydrogen peroxide'}],
    [{'TEXT': {'REGEX': '\\d,\\d'}}]
]

example_text = [
    'hydrogen peroxide',
    '2,3-butanediol'
]

ruler.add_patterns([{'label': 'PATTERN', 'pattern': pattern} for pattern in patterns])

for sent in example_text:
    print(sent)
    doc = model(sent)
    print([t.text for t in doc])
    #print([(t, t.whitespace_) for t in doc])
    #matches = matcher(doc, as_spans=True)
    print([t for t in doc.spans['ruler']])
```
