```python
import spacy
from spacy.pipeline import EntityRuler
from spacy.tokens import Doc
import re
import json
```

```python
model = spacy.load('en_core_web_sm')
ruler = model.add_pipe('entity_ruler', before='ner')

compound_regex_patterns = [
    [{'TEXT': {'REGEX': '(?<!ase)$'}}]
]

ruler.add_patterns([{'label': 'ENTITY', 'pattern': pattern} for pattern in compound_regex_patterns])
```

```python
tests = [
    'glucose oxidase is an enzyme',
    'ammonia, urea and fructose',
    'i been married a long time ago where did she come from where did she go',
    '2,3-butanase 2,3-butanol'
]

for test in tests:
    print(test)
    doc = model(test)

    print([(t.label_, t) for t in doc.ents])
```
