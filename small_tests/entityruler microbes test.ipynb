```python
#note to self: spacy DOES NOT work with python 3.14!!!!!!
import spacy
from spacy.pipeline import EntityRuler
from spacy import displacy

import json
from pypdf import PdfReader as pr
import os

with open('..\\species_names.json') as f:
    microbial_species = json.load(f)
    
with open('..\\genus_names.json') as f:
    microbial_genera = json.load(f)
    
with open('..\\microbe_names.json') as f:
    microbes = json.load(f)
```

```python
print(microbial_species[0])

model = spacy.load('en_core_web_sm')

model.remove_pipe('ner')
ner_microbe = model.add_pipe('ner', 'microbe_ner', after='lemmatizer')
regular_ner = model.add_pipe('ner', 'regular_ner', after='microbe_ner')

ner_microbe.add_label('MICROBE')

labels = ['GENERA', 'SPECIES', 'PERCENT', 'QUANTITY', 'TIME', 'CARDINAL', 'DATE', 'ORDINAL']
for label in labels:
    regular_ner.add_label(label)
    
ruler = model.add_pipe('entity_ruler', after='lemmatizer', config={'overwrite_ents': True})

print(regular_ner.labels)
print(model.pipe_names)
```
