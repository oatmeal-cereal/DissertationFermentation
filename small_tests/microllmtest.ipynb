```python
from transformers import AutoTokenizer as autok
from transformers import BertForTokenClassification as bftc
from transformers import 
from transformers import pipeline
import spacy
import os

os.environ['HF_HUB_DISABLE_SYMLINKS_WARNING'] = '1'
```

```python
biobert = 'dmis-lab/biobert-v1.1'
llmner = 'timo-10000/LLM-NER'
chatglm = 'zai-org/GLM-4.7'
scibert = 'allenai/scibert_scivocab_uncased'

tokenizer = autok.from_pretrained(llmner)
model = bftc.from_pretrained(llmner)
nlp = pipeline('ner', model=model, tokenizer=tokenizer)
```

```python
with open('list_files\\all sentences.txt', 'r', encoding='utf-8') as f:
    sentences = f.readlines()
```

```python
for line in sentences:
    print(line)
    entities = nlp(line)
    for entity in entities:
        print(entity.get('entity'))
        start = int(entity.get('start'))
        end = int(entity.get('end'))
        print(line[start:end])
    print()
```
