```python
import pubchempy as pcp
```

```python
test_compounds = ['lactic acid', 'propionic acid', 'propanoic acid', 'ethanol', 'bioethanol', 'riboflavin', 'vitamin B2', 'glucose', 'fructose', 'xylose', '2,3-butanediol', 'benzoate', 'formic acid', 'methanoic acid']

test_compounds2 = ['2-ketobutyric acid', 'glutamic acid', 'tricarboxylic acid', 'isoflavone aglycones', 'gluconic acid', 'acetic acid', 'lysine', '2-furfuryl methanol', 'oxobutanoic acid']

test_compounds3 = ['silver/silver chloride', 'intermediary']

for compound in test_compounds3:
    comps = pcp.get_compounds(compound, 'name')
    print([comp.iupac_name for comp in comps])
    cids = pcp.get_cids(compound, 'name', 'compound', list_return='flat')
    if cids:
        print(pcp.Compound.from_cid(cids[0]).synonyms[0])
    if comps:
        print(comps[0].synonyms[0])
    print()
    
#propionic and propanoic acid should return the same thing
#so should ethanol and bioethanol
```
