```python
import spacy
from spacy import displacy
from spacy.tokens import Span, Doc
```

```python
nlp = spacy.blank('en')
```

```python
with open('..\\list_files\\all sentences.txt', 'r', encoding='utf-8') as f:
    sentences = f.readlines()
    
with open('..\\list_files\\microllm_ents.txt', 'r', encoding='utf-8') as f:
    all_ents = f.readlines()
```

```python
def merge_overlapping_spans(spans):
    prev_start = 0
    prev_end = 0
    prev_label = ''
    new_spans = []
    for e_start, e_end, label in spans:
        if e_start < end:
            if new_spans:
                new_spans.pop()
            new_spans.append((prev_start, e_end, prev_label))
        else:
            new_spans.append((e_start, e_end, label))
        prev_start = e_start
        prev_end = e_end
        prev_label = label
    return new_spans

for i, line in enumerate(sentences):
    print(i)
    doc = nlp.make_doc(line)
    line_ents = all_ents[i].split('||')
    spans = []
    print_ents = []
    for ent in line_ents:
        if ent == '\n' or not ent:
            continue
        e_spl = ent.split(',,')
        start = int(e_spl[0])
        end = int(e_spl[1])
        print_ents.append((start, end, line[start:end]))
        spans.append((start, end, e_spl[2]))
    if not spans:
        continue
    try:
        doc_spans = [doc.char_span(start, end, label=label, alignment_mode='expand') for start, end, label in spans]
        doc.ents = doc_spans
        displacy.render(doc, style='ent', jupyter=True, minify=True)
    except:
        print(spans)
        print(print_ents)
        print(line)
```
