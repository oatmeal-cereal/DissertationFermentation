```python
import spacy
from spacy import displacy
from spacy.tokens import Span

nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp.make_doc("The thing is this")
doc.ents = [doc.char_span(0, 3, label='THE'), doc.char_span(4, 9, label='THING')]
displacy.render(doc, style='ent', jupyter=True, minify=True, page=True)
```
