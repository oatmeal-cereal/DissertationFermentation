```python
import json
import re

compounds = set()

compound_endings = set()

ending_no = 5

with open('list_files\\web scraped organic compounds list.txt', 'r', encoding='utf-8') as f:
    for line in f.readlines():
        line = line.strip()
        stuff = line.split(' — ')
        if len(stuff) == 1:
            stuff = line.split(' – ')
        compounds.add(stuff[0].lower())
        compound_parts = re.split('[\\s-]', stuff[0])
        ending = compound_parts[-1][-ending_no:]
        if ending.isalpha() and len(ending) >= 3 and ending.islower():
            compound_endings.add(ending.lower())
            
        if ending[-3:] == 'ine':
            compounds.add(compound_parts[-1][:-1].lower())
        
compound_endings = compound_endings - set(['blue'[-ending_no:], 'purple'[-ending_no:], 'indigo'[-ending_no:], 'violet'[-ending_no:], 'orange'[-ending_no:], 'red'])

print(len(compounds))

with open(f'list_files\\compound_{ending_no}_endings.json', 'w', encoding='utf-8') as f:
    json.dump(list(compound_endings), f)

with open('list_files\\compounds_list.json', 'w', encoding='utf-8') as f:
    json.dump(list(compounds), f)
```
