```python
import spacy
from spacy.tokens import Doc
from spacy import displacy
import os
import json

path_to_custom_model = os.path.normpath('C:\\Users\\jace\\Documents\\assignments\\dissertation DLC content\\custom_ner')

model = spacy.load(path_to_custom_model)
```

```python
paper_name1 = '10.1016-j.fm.2019.02.007'
paper_name2 = '10.1016-j.foodchem.2022.134488'

filepath = f'..\\dissertation DLC content\\fermentation_papers_txt\\{paper_name1}.txt'

with open(filepath, 'r') as f:
    paper_text = f.read()

my_labels = ['MICROBE_NAME', 'SIMPLE_CHEMICAL']

paper_doc = model(paper_text)

new_paper_ents = []

for ent in paper_doc.ents:
    if ent.label_ in my_labels:
        new_paper_ents.append(ent)

paper_doc.ents = new_paper_ents

displacy.render(paper_doc, style='ent')
```
