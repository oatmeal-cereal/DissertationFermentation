```python
import pandas
import sqlite3

filepath = '..\\dissertation DLC content\\MiFoDB_gamma_v1_AllReferences.csv'

database = pandas.read_csv(filepath, encoding='latin')

print(database.info())
```

```python
conn = sqlite3.connect('fermentedfoods.db') # Connecting to the database
cursor = conn.cursor() # Object to run queries

columns_with_types = ", ".join([f"{col.replace(' ', '_')} TEXT" for col in database.columns])
cursor.execute("DROP TABLE fermentedfoods")
create_table_query = f"""CREATE TABLE IF NOT EXISTS "fermentedfoods" (genus TEXT, species TEXT, sample TEXT, substrate TEXT);"""

cursor.execute(create_table_query)
```

```python
for index, row in database.iterrows():
    values = ", ".join([f'"{str(row["genus"])}"', f'"{str(row["species"])}"', f'"{str(row["sample"])}"', f'"{str(row["substrate"])}"'])
    insert_sql = f"INSERT INTO fermentedfoods (genus, species, sample, substrate) VALUES ({values})"
    cursor.execute(insert_sql)

cursor.execute('pragma table_info(fermentedfoods);')
cursor.fetchall()
```

```python
cursor.execute('SELECT COUNT(*) FROM fermentedfoods')
cursor.fetchall()

conn.commit()
conn.close()
```
